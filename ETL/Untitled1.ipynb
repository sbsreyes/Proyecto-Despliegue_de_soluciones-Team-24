```python
import pandas as pd

df_alt = pd.read_csv('municipios_colombia_altitud.csv')
print(f"Shape: {df_alt.shape}")
print(f"Columnas: {df_alt.columns.tolist()}")
print(f"\nPrimeras filas:")
print(df_alt.head())
print(f"\nDtypes:")
print(df_alt.dtypes)
print(f"\nNulos:")
print(df_alt.isna().sum())

# Si el notebook usa solo estas 4 columnas, nos quedamos con esas
df_alt_clean = df_alt[['latitud','longitud','altitud_msnm','cat_altitud']].dropna()

print(f"\nDespués de limpiar: {len(df_alt_clean):,} puntos")
print(f"\nDistribución cat_altitud:")
print(df_alt_clean['cat_altitud'].value_counts())
print(f"\nRango altitud: [{df_alt_clean['altitud_msnm'].min()}, {df_alt_clean['altitud_msnm'].max()}] msnm")

df_alt_clean.to_parquet("altitud.parquet", compression='gzip')
print("\n✅ altitud.parquet generado")
```

```python
import pandas as pd
import gdown

# Descargar
gdown.download('https://drive.google.com/uc?id=1S76FGFaw6Sywj-sfOWcl6UqOCsbb1M2i',
               'dataset_dengue_completo.csv', quiet=False)

# Cargar y validar
df = pd.read_csv('dataset_dengue_completo.csv')
df['cod_municipio'] = df['cod_municipio'].astype(str).str.zfill(5)

print(f"Shape: {df.shape}")
print(f"Memoria: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB\n")

print(f"Años: {sorted(df['anio'].unique())}")
print(f"Municipios únicos: {df['cod_municipio'].nunique()}\n")

print(f"Columnas ({len(df.columns)}):")
print(df.columns.tolist())

print(f"\nDtypes:")
print(df.dtypes)

print(f"\nNulos por columna:")
print(df.isna().sum()[df.isna().sum() > 0])  # solo columnas con nulos

print(f"\nMuestra (3 primeras filas):")
print(df.head(3))

print(f"\nEstadísticas numéricas:")
print(df.describe())

# Guardar parquet
df.to_parquet('dataset_dengue_completo.parquet', compression='gzip')
print(f"\n✅ Parquet generado: dataset_dengue_completo.parquet")
```

```python
import pandas as pd

df = pd.read_parquet('dataset_dengue_completo.parquet')

print(f"Shape: {df.shape}")
print(f"Memoria: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB\n")

print(f"=== COLUMNAS Y TIPOS ===")
print(df.dtypes)

print(f"\n=== NULOS POR COLUMNA ===")
nulos = df.isna().sum()
print(nulos[nulos > 0] if nulos.sum() > 0 else "Sin nulos")

print(f"\n=== AÑOS Y MUNICIPIOS ===")
print(f"Años: {sorted(df['anio'].unique())}")
print(f"Municipios únicos: {df['cod_municipio'].nunique()}")

print(f"\n=== MUESTRA (3 FILAS) ===")
print(df.head(3))

print(f"\n=== ESTADÍSTICAS NUMÉRICAS ===")
print(df.describe())

# Si hay columna target/alerta
target_cols = [c for c in df.columns if 'alerta' in c.lower() or 'target' in c.lower() or 'label' in c.lower()]
if target_cols:
    print(f"\n=== TARGET ({target_cols}) ===")
    for col in target_cols:
        print(f"{col}: {df[col].value_counts().to_dict()}")
```
